# Preprocessing Data (employment_length_in_years)

```python
# Preprocessing Data
# str.replace(<string to be replaced>, <new replacement string>)
# replaces all string of a Series with new strings

def employment_length_in_years():
    
    import pandas as pd
    import numpy as np
    
    df = pd.read_csv('lc_loan.csv')
    lc_loan = df.copy()
    
    lc_loan['emp_length'].unique()
    lc_loan['emp_length_int'] = lc_loan['emp_length'].str.replace('\+ years','')
    lc_loan['emp_length_int'] = lc_loan['emp_length_int'].str.replace('< 1 year',str(0))
    lc_loan['emp_length_int'] = lc_loan['emp_length_int'].str.replace('n/a',str(0))
    lc_loan['emp_length_int'] = lc_loan['emp_length_int'].str.replace('years','')
    lc_loan['emp_length_int'] = lc_loan['emp_length_int'].str.replace(' year','')
    lc_loan['emp_length_int'].unique()
    
    
    # type of data
    type(lc_loan['emp_length_int'][0])
    
    # Now they are dtype: float64
    lc_loan['emp_length_int'] = pd.to_numeric(lc_loan['emp_length_int'])
    
    return lc_loan['emp_length_int']

employment_length_in_years()
```

```python
# Preprocessing few discrete variables
# Biological sex variable
# Male (1) and Female (0)

def dummies_values():
    
    import pandas as pd
    import numpy as np
    df = pd.read_csv('lc_loan.csv')
    lc_loan = df.copy()
    print(lc_loan.info())
    print('   ')
    
    
    
    # pd.get_dummies(Series) creates dummy variables for a given  categorical variable
    pd.get_dummies(lc_loan['grade'])
    print(pd.get_dummies(lc_loan['grade']))
    
    
    
    # pd.get_dummies(Series, [prefix], [prefix_sep]) creates dummy variables for a given categorical variable; we can specify the prefix to the name of the dummy variables using the prefix and prefix_sep arguments
    pd.get_dummies(lc_loan['grade'], prefix = 'grade', prefix_sep = ':')
    print(pd.get_dummies(lc_loan['grade'], prefix = 'grade', prefix_sep = ':'))
    
    loan_data_dummies = [pd.get_dummies(lc_loan['grade'], prefix = 'grade', prefix_sep = ':'),
             pd.get_dummies(lc_loan['sub_grade'], prefix = 'sub_grade', prefix_sep = ':'),
             pd.get_dummies(lc_loan['home_ownership'], prefix = 'home_ownership', prefix_sep = ':'),
             pd.get_dummies(lc_loan['verification_status'], prefix = 'verification_status', prefix_sep = ':'),
             pd.get_dummies(lc_loan['loan_status'], prefix = 'loan_status', prefix_sep = ':'),
             pd.get_dummies(lc_loan['purpose'], prefix = 'purpose', prefix_sep = ':'),
             pd.get_dummies(lc_loan['addr_state'], prefix = 'addr_state', prefix_sep = ':'),
             pd.get_dummies(lc_loan['initial_list_status'], prefix = 'initial_list_status', prefix_sep = ':'),
            ]
    loan_data_dummies = pd.concat(loan_data_dummies, axis = 1)
    type(loan_data_dummies)
    
    loan_data_dummies
    
    # axis = 0 , refers to rows (observation);
    # aixs = 1 , refers to columns (variables)
    loan_Data = pd.concat([loan_data_dummies, lc_loan], axis=1)
    
    loan_Data.columns.values
    
    return loan_Data.columns.values

dummies_values()
```
